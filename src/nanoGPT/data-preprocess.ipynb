```python
import json
import tiktoken
import numpy as np
import os
```

```python
enc = tiktoken.encoding_for_model("gpt-4")
```

```python
data = json.load(open('poet.tang.0.json'))
```

```python
print(len(data))
```

```python
print(data[0])
```

```python
train, val = data[:900], data[900:]
```

```python
def process(poem):
    # ids = enc.encode_ordinary(example['text']) # encode_ordinary ignores any special tokens
    # ids.append(enc.eot_token) # add the end of text token, e.g. 50256 for gpt2 bpe
    # out = {'ids': ids, 'len': len(ids)}
    text = poem['title'] + '\n' + poem['author'] + '\n' + '\n'.join(poem['paragraphs'])
    ids = enc.encode(text)
    ids.append(enc.eot_token)
    return ids
```

```python
train_tk = map(process, train)
```

```python
val_tk = map(process, val)
```

```python
print(list(val_tk))
```

```python
a=[1,2]
b=[3,4,5]
print(a+b)
ll = [a,b]
c = []
for ia in ll:
    c=c+ia
print(c)
```

```python

# encode both to integers
train_ids = []
for tids in list(train_tk):
    train_ids=train_ids+tids
val_ids = [] 
for vids in list(val_tk):
    val_ids=val_ids+vids
print(f"train has {len(train_ids):,} tokens")
print(f"val has {len(val_ids):,} tokens")
```

```python
print(f"train has {len(train_ids):,} tokens")
print(f"val has {len(val_ids):,} tokens")
```

```python

# export to bin files
train_ids = np.array(train_ids).astype(np.uint16)
val_ids = np.array(val_ids).astype(np.uint16)
train_ids.tofile('train.bin')
val_ids.tofile('val.bin')
```
